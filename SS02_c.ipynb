## 複數弦波訊號

複數弦波訊號 $x(t)$ 係由 [複數指數函數](https://en.wikipedia.org/wiki/Exponential_function#Complex_plane) 所定義

\begin{equation}
x(t) = e^{s t} 
\end{equation}

其中 $s = \alpha + j \omega_0$. $\alpha, \omega_0 \in \mathbb{R}$ 與 $j$ 為虛數符號 $(j^2=-1)$. 利用尤拉公式[Euler's formula](https://en.wikipedia.org/wiki/Euler's_formula) 可將上述 $x(t)$ 訊號改寫成

\begin{equation}
x(t) = e^{(\alpha + j \omega_0) t} = e^{\alpha t} \cos(\omega_0 t) + j e^{\alpha t} \sin(\omega_0 t)
\end{equation}

實部與虛部均為指數乘上弦波。$\omega_0 = 2 \pi f_0$.

對於 $t>0$, 指數項   $e^{\alpha t}$ 決定訊號強弱

* $\alpha < 0$, 訊號指數遞減
* $\alpha = 0$, 訊號強弱不變
* $\alpha > 0$, 訊號指數遞增

至於 $t<0$ 則得到與 $t>0$ 相反之結果

## 一般化情況: ($s=\alpha+j\omega_0$)

底下程式使用 Python 模組 [SymPy](http://docs.sympy.org/latest/index.html) 定義 複指數訊號 $x(t)=e^{jst}=e^{j(\alpha+j\omega_0)t}$。變數 $t$, $\omega_0$ and $\alpha$ 均為實數變數。

```python
import sympy as sym
sym.init_printing()

t, alpha, omega0 = sym.symbols('t alpha omega', real=True)
s = alpha + 1j*omega0
x = sym.exp(s*t)
x
```

## Case 1: $s$ 為實數 ($s=\alpha$)

```python
# 設定 alpha 與 sigma 數值
# alpha<0
y = x.subs({omega0: 0, alpha: -.5})  # alpha=-0.5 
sym.plot(sym.re(y), (t, -1, 2*sym.pi), ylabel=r'$e^{st}$',title=r'$\alpha<0$')

# alpha>0
y = x.subs({omega0:0, alpha:0.5}) #alpha=0.5
p=sym.plot(sym.re(y), (t, -1,2*sym.pi), ylabel=r'Im{$e^{st}$}',title=r'$\alpha>0$');
```

## Case 2: s為純虛數 ($s=j\omega_0$)

```python
# 設定 alpha 與 sigma 數值
y = x.subs({omega0: 10, alpha: 0})  # alpha=0, omega=10

sym.plot(sym.re(y), (t, 0, 2*sym.pi), ylabel=r'Re{$e^{st}$}', title='Real part')
sym.plot(sym.im(y), (t, 0, 2*sym.pi), ylabel=r'Im{$e^{st}$}', title='Imaginary part');
```

## Case 3: 一般化情況 ($s=\alpha+j\omega_0$)

### $\alpha<0$, 震盪遞減

```python
# 設定 alpha 與 sigma 數值
y = x.subs({omega0: 10, alpha: -0.2})  # alpha=-0.2, omega=10

sym.plot(sym.re(y), (t, 0, 2*sym.pi), ylabel=r'Re{$e^{st}$}',title=r'$\alpha<0$')
sym.plot(sym.im(y), (t, 0, 2*sym.pi), ylabel=r'Im{$e^{st}$}',title=r'$\alpha<0$');
```

### $\alpha>0$, 震盪遞增

```python
# 設定 alpha 與 sigma 數值
y = x.subs({omega0: 10, alpha: 0.2})  # alpha=0.2, omega=10

sym.plot(sym.re(y), (t, 0, 2*sym.pi), ylabel=r'Re{$e^{st}$}',title=r'$\alpha>0$')
sym.plot(sym.im(y), (t, 0, 2*sym.pi), ylabel=r'Im{$e^{st}$}',title=r'$\alpha>0$');
```
